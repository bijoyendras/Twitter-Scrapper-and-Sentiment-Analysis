```python

df = pd.read_excel("output.xlsx")
```

```python

stemmer = PorterStemmer()
stop_words = set(stopwords.words('english'))

def clean_text(text):
    # convert text to lowercase
    text = text.lower()
    # remove emojis
    text = re.sub(r'[^\x00-\x7F]+','', text)
    # handle twitter names
    text = re.sub(r'@[A-Za-z0-9]+','',text)
    # remove web links
    text = re.sub(r'https?:\/\/\S+','',text)
    # remove punctuation
    text = re.sub(r'[^\w\s]','',text)
    # remove "rt" from text
    text = re.sub(r'rt\b', '', text)
    
    # tokenize the text
    words = word_tokenize(text)
    # stem the words and remove stop words
    stemmed_words = [stemmer.stem(word) for word in words if word.lower() not in stop_words]
    # join the stemmed words to form a cleaned text
    cleaned_text = ' '.join(stemmed_words)
    return cleaned_text
```

```python

df['cleaned_text'] = df['text'].apply(clean_text)
```

```python
# export the cleaned_text column into a .csv file
df.to_csv('cleaned_text.csv', index=False, header=False)

#read the .csv file into a single string
with open('cleaned_text.csv', 'r') as file:
  text = file.read().replace('\n',' ')

# create a Counter object to count the occurrences of each word
word_counts = Counter(text.split())

# keep only the top 25 words
top_words = dict(word_counts.most_common(25))

# create a wordcloud using only the top 25 words
wordcloud = WordCloud(width = 800, height = 800, 
                background_color ='white', 
                stopwords = stop_words, 
                min_font_size = 10).generate_from_frequencies(top_words)

#create a list of words
words = text.split()

#create a frequency counter
word_freq = Counter(words)

#select the top 25 words
top_25_words = dict(word_freq.most_common(25))

#create a wordcloud
wordcloud = WordCloud(width = 800, height = 800,
background_color ='white',
stopwords = stop_words,
min_font_size = 10).generate_from_frequencies(top_25_words)

#plot the wordcloud
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud)
plt.axis("off")
plt.tight_layout(pad = 0)

plt.show()
```

```python

# import the library for sentiment analysis
from textblob import TextBlob

# create a new column 'sentiment' with sentiment values
df["sentiment"] = df["cleaned_text"].apply(lambda x: TextBlob(x).sentiment.polarity)

# plot the sentiment distribution
sns.distplot(df["sentiment"])
plt.show()
```

```python

# create a function to classify sentiment as positive, negative, or neutral
def classify_sentiment(polarity):
    if polarity > 0:
        return "positive"
    elif polarity < 0:
        return "negative"
    else:
        return "neutral"

# apply the function to the sentiment column to create a new column 'sentiment_class'
df["sentiment_class"] = df["sentiment"].apply(classify_sentiment)

# plot the sentiment distribution
sns.countplot(x='sentiment_class', data=df)
plt.show()
```
